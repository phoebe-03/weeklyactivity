```python
#Exercise 2 
#simple custom function to determine whether a given integer is odd or even number.

def odd_even(x): 
    if isinstance(x, int):
        if(x % 2 == 0 and x != 0):
            print (f"{x} is integer")
        elif (x % 2 == 1):
            print (f"{x} is odd")
        elif (x == 0):
              print (f"{x} is integer")
    else: 
        print (f"{x} is not integer")
```

```python
odd_even(8.9)
```

```python
odd_even(1)
```

```python
#Exercise 3
#Numpy array
arr_a = np.arange(1,100000)
arr_b = np.arange(6,100005)
result = arr_a * arr_b
print(result)
```

```python
list_a = list(range(1,100000))
list_b = list(range(6,100005))
length = len (list_a)
result_list = [list_a[i]* list_b[i] for i in range(length)]
print(result_list) 
```

```python
#Exercise 5
import numpy as np
import matplotlib.pyplot as plt

# Create a black image
black = np.zeros((200, 200), dtype=np.uint8)

# Create a figure and axis for the circle
fig, ax = plt.subplots()

# Add a white circle to the axis
circle = plt.Circle((100, 100), 30 , color='white')
ax.imshow(black, cmap='gray', vmin=0, vmax=255)
ax.set_aspect(1)
ax.add_artist(circle)
ax.patch.set_facecolor('black')  # Set background color of the axis

# Show the plot
plt.show()
```
